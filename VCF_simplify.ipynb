```python
vcf = "merge_result.vcf"

def getColNames(vcf):
    colNames = []
    with open(vcf, "r") as f:
        for line in f:
            if line.startswith("#CHROM"):
                for column in line.split("\t"):
                    colNames.append(column)
        #To remove the \n at the end of the last column's name
        #colNames = colNames[:-1]
        #colNames.append(colNames[-1].split("\n")[0]) 
    return colNames

print(getColNames(vcf))
```

```python
def vcfToDf(vcf):
    df = pd.read_csv(vcf, sep='\t', comment='#', names=getColNames(vcf), header=0)
    return df

vcfToDf(vcf)
```

```python
def vcfSummary(df):
    df["SVTYPE"]=df["INFO"].str.split('SVTYPE=', expand=True)[1]
    df["SVLEN"]=df["INFO"].str.split('SVLEN=', expand=True)[1].str.split(';',expand=True)[0]
    
    for column in ['P15-8.trimed1000', 'P30-8.trimed1000', 'P50-8.trimed1000\n'] :
        list_DR = []
        list_DV = []
        
        for format, values in zip(df["FORMAT"], df[column]):
            list_format = format.split(':')
            index_DR = list_format.index("DR")
            index_DV = list_format.index("DV")
            list_values = values.split(':')
            
            if len(list_values)==4 :
                list_DR.append(values.split(':')[index_DR])
                list_DV.append(values.split(':')[index_DV])
            else :
                list_DR.append(".")
                list_DV.append(".")
        
        passage = column.split(".")[0]
        DR = passage+"_DR"
        DV = passage+"_DV"
        df[DR] = list_DR
        df[DV] = list_DV
        df[DR] = pd.to_numeric(df[DR], errors="coerce").fillna(0).astype(int) # transform into int; if string then changed to nan
        df[DV] = pd.to_numeric(df[DV], errors="coerce").fillna(0).astype(int) # transform into int; if string then changed to nan
        df[passage+"_AF"] = (df[DV])/(df[DV] + df[DR])
        df[passage+"_AF"] = df[passage+"_AF"].fillna('.')
        
    df = df. loc[:, ['#CHROM', 'POS', 'SVTYPE', 'SVLEN', 'P15-8_DR', 'P15-8_DV', 'P15-8_AF','P30-8_DR', 'P30-8_DV', 'P30-8_AF', 'P50-8_DR', 'P50-8_DV', 'P50-8_AF']]
    
    return df

vcfSummary(vcfToDf(vcf))
```

```python
df=vcfSummary(vcfToDf(vcf))
df.to_csv(r'merge_result_simplified.vcf', header=True, index=None, sep='\t')
```
